```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
# Load the dataset
dataset = pd.read_excel('images-finished.xlsx')
dataset = dataset[:100]
```

```python
dataset
```

```python
dataset['Log_Count'] = dataset['actual count'].apply(lambda x: np.log(x))
```

```python
# Split the dataset into features (X) and target variable (y)
X = dataset[['Front','Close','Log_Count']]
y = dataset['hit rate']
```

```python
X
```

```python
y
```

```python
# Create the linear regression model and fit it to the data
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
# Make predictions on the same data
y_pred = regressor.predict(X)
```

```python
# Evaluate the model using R^2 score
r2 = r2_score(y, y_pred)
print("R-squared value: ", r2)
```

```python
# Add a constant column to X
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary table
model.summary()
```
