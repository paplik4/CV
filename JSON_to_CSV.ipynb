```python
import pandas as pd 
import json
import os
```

# Test data

```python
# Initialize empty DataFrame
df = pd.DataFrame()

json_path = '/Users/dkAndHus/Desktop/barcode_data/bar_code/barcode_generator/data/test/'

# Loop over JSON files in directory
for filename in os.listdir(json_path):
    if filename.endswith('.json'):
        # Load JSON file into Python object
        with open('/Users/dkAndHus/Desktop/barcode_data/bar_code/barcode_generator/data/test/' + filename) as f:
            data = json.load(f)

        # Create pandas DataFrame from JSON data
        df_temp = pd.json_normalize(data)

        # Extract 'ymax' value for each object in the DataFrame
        xmin_list = []
        ymin_list = []
        xmax_list = []
        ymax_list = []


        for obj in data:
            xmin = obj['bbox']['xmin']
            ymin = obj['bbox']['ymin']
            xmax = obj['bbox']['xmax']
            ymax = obj['bbox']['ymax']
            xmin_list.append(xmin)
            ymin_list.append(ymin)
            xmax_list.append(xmax)
            ymax_list.append(ymax)

        # Add 'ymax' values to the DataFrame
        df_temp['filename'] = filename.replace('.json', '.png')
        df_temp['xmin'] = xmin_list
        df_temp['ymin'] = ymin_list
        df_temp['xmax'] = xmax_list
        df_temp['ymax'] = ymax_list
        df_temp['class'] = 'barcode'

        # Append DataFrame to the main DataFrame
        df = pd.concat([df, df_temp], ignore_index=True)
df.head()
```

```python
df_fin = df[['filename', 'width','height','class','xmin','ymin','xmax','ymax']]
df_fin
```

```python
# Change this code to the proper path and all the other lines
os.makedirs('/Users/dkAndHus/Desktop/barcode_data/bar_code/barcode_generator/data', exist_ok=True)  
df_fin.to_csv('/Users/dkAndHus/Desktop/barcode_data/bar_code/barcode_generator/data/test_data.csv')
```
